### Markov decision process

This week's methods are all built to solve __M__arkov __D__ecision __P__rocesses. In the broadest sense, an MDP is defined by how it changes states and how rewards are computed.

### Value Iteration

Now let's build something to solve this MDP. The simplest algorithm so far is __V__alue __I__teration

Here's the pseudo-code for VI:

---

`1.` Initialize $V^{(0)}(s)=0$, for all $s$

`2.` For $i=0, 1, 2, \dots$
 
`3.` $ \quad V_{(i+1)}(s) = \max_a \sum_{s'} P(s' | s,a) \cdot [ r(s,a,s') + \gamma V_{i}(s')]$, for all $s$

---

First, let's write a function to compute the state-action value function $Q^{\pi}$, defined as follows

$$Q_i(s, a) = \sum_{s'} P(s' | s,a) \cdot [ r(s,a,s') + \gamma V_{i}(s')]$$


```python
def get_action_value(mdp, state_values, state, action, gamma):
    """ Computes Q(s,a) as in formula above """
    q = 0
    for next_state, prob in mdp.get_next_states(state, action).items():
        r = mdp.get_reward(state, action, next_state)
        q += prob * (r + gamma * state_values[next_state])
    return q
```

```python
import numpy as np
test_Vs = {s: i for i, s in enumerate(sorted(mdp.get_all_states()))}
assert np.isclose(get_action_value(mdp, test_Vs, 's2', 'a1', 0.9), 0.69)
assert np.isclose(get_action_value(mdp, test_Vs, 's1', 'a0', 0.9), 3.95)
```

Using $Q(s,a)$ we can now define the "next" V(s) for value iteration.
 $$V_{(i+1)}(s) = \max_a \sum_{s'} P(s' | s,a) \cdot [ r(s,a,s') + \gamma V_{i}(s')] = \max_a Q_i(s,a)$$

```python
def get_new_state_value(mdp, state_values, state, gamma):
    """ Computes next V(s) as in formula above. Please do not change state_values in process. """
    if mdp.is_terminal(state):
        return 0

    actions = mdp.get_possible_actions(state)
    return max(get_action_value(mdp, state_values, state, a, gamma) for a in actions)
```

```python
test_Vs_copy = dict(test_Vs)
assert np.isclose(get_new_state_value(mdp, test_Vs, 's0', 0.9), 1.8)
assert np.isclose(get_new_state_value(mdp, test_Vs, 's2', 0.9), 1.08)
assert np.isclose(get_new_state_value(mdp, {'s0': -1e10, 's1': 0, 's2': -2e10}, 's0', 0.9), -13500000000.0), \
    "Please ensure that you handle negative Q-values of arbitrary magnitude correctly"
assert test_Vs == test_Vs_copy, "Please do not change state_values in get_new_state_value"
```

Finally, let's combine everything we wrote into a working value iteration algo.

```python
# parameters
gamma = 0.9            # discount for MDP
num_iter = 100         # maximum iterations, excluding initialization
# stop VI if new values are this close to old values (or closer)
min_difference = 0.001

# initialize V(s)
state_values = {s: 0 for s in mdp.get_all_states()}

if has_graphviz:
    display(plot_graph_with_state_values(mdp, state_values))

for i in range(num_iter):

    # Compute new state values using the functions you defined above.
    # It must be a dict {state : float V_new(state)}
    new_state_values = {s: get_new_state_value(mdp, state_values, s, gamma)
                    for s in mdp.get_all_states()}

    assert isinstance(new_state_values, dict)

    # Compute difference
    diff = max(abs(new_state_values[s] - state_values[s])
               for s in mdp.get_all_states())
    print("iter %4i   |   diff: %6.5f   |   " % (i, diff), end="")
    print('   '.join("V(%s) = %.3f" % (s, v) for s, v in state_values.items()))
    state_values = new_state_values

    if diff < min_difference:
        print("Terminated")
        break
```

Now let's use those $V^{*}(s)$ to find optimal actions in each state

 $$\pi^*(s) = argmax_a \sum_{s'} P(s' | s,a) \cdot [ r(s,a,s') + \gamma V_{i}(s')] = argmax_a Q_i(s,a)$$
 
The only difference vs V(s) is that here we take not max but argmax: find action such with maximum Q(s,a).

```python
def get_optimal_action(mdp, state_values, state, gamma=0.9):
    """ Finds optimal action using formula above. """
    if mdp.is_terminal(state):
        return None

    actions = mdp.get_possible_actions(state)
    return max(actions, key=lambda a: get_action_value(mdp, state_values, state, a, gamma))
```

```python
assert get_optimal_action(mdp, state_values, 's0', gamma) == 'a1'
assert get_optimal_action(mdp, state_values, 's1', gamma) == 'a0'
assert get_optimal_action(mdp, state_values, 's2', gamma) == 'a1'

assert get_optimal_action(mdp, {'s0': -1e10, 's1': 0, 's2': -2e10}, 's0', 0.9) == 'a0', \
    "Please ensure that you handle negative Q-values of arbitrary magnitude correctly"
assert get_optimal_action(mdp, {'s0': -2e10, 's1': 0, 's2': -1e10}, 's0', 0.9) == 'a1', \
    "Please ensure that you handle negative Q-values of arbitrary magnitude correctly"
```

```python
def value_iteration(mdp, state_values=None, gamma=0.9, num_iter=1000, min_difference=1e-5):
    """ performs num_iter value iteration steps starting from state_values. Same as before but in a function """
    state_values = state_values or {s: 0 for s in mdp.get_all_states()}
    for i in range(num_iter):

        # Compute new state values using the functions you defined above. It must be a dict {state : new_V(state)}
        new_state_values = {s: get_new_state_value(mdp, state_values, s, gamma)
                            for s in mdp.get_all_states()}

        assert isinstance(new_state_values, dict)

        # Compute difference
        diff = max(abs(new_state_values[s] - state_values[s])
                   for s in mdp.get_all_states())

        print("iter %4i   |   diff: %6.5f   |   V(start): %.3f " %
              (i, diff, new_state_values[mdp._initial_state]))

        state_values = new_state_values
        if diff < min_difference:
            break

    return state_values
```

### Value iteration convervence proof (1 pts)
**Note:** Assume that $\mathcal{S}, \mathcal{A}$ are finite.

Update of value function in value iteration can be rewritten in a form of Bellman operator:

$$(TV)(s) = \max_{a \in \mathcal{A}}\mathbb{E}\left[ r_{t+1} + \gamma V(s_{t+1}) | s_t = s, a_t = a\right]$$

Value iteration algorithm with Bellman operator:

---
&nbsp;&nbsp; Initialize $V_0$

&nbsp;&nbsp; **for** $k = 0,1,2,...$ **do**

&nbsp;&nbsp;&nbsp;&nbsp; $V_{k+1} \leftarrow TV_k$

&nbsp;&nbsp;**end for**

---

In [lecture](https://docs.google.com/presentation/d/1lz2oIUTvd2MHWKEQSH8hquS66oe4MZ_eRvVViZs2uuE/edit#slide=id.g4fd6bae29e_2_4) we established contraction property of bellman operator:

$$
||TV - TU||_{\infty} \le \gamma ||V - U||_{\infty}
$$

For all $V, U$

Using contraction property of Bellman operator, Banach fixed-point theorem and Bellman equations prove that value function converges to $V^*$ in value $iteration$

**Доказательство.**

Ключевое свойство, на которое мы опираемся, — сжимаемость оператора $T$: для любых $V, U$

$$\|TV - TU\|_\infty \le \gamma\,\|V - U\|_\infty, \qquad \gamma \in [0,1).$$

Сжимающее отображение на полном пространстве по теореме Банаха обладает ровно одной неподвижной точкой; обозначим её $V_f$, так что $V_f = TV_f$. Но равенство $V = TV$ — это в точности уравнение оптимальности Беллмана, единственным решением которого служит $V^*$. Значит, $V_f = V^*$, и других неподвижных точек у $T$ нет.

Посмотрим теперь, как меняется отклонение $\|V_k - V^*\|_\infty$ вдоль итераций $V_{k+1} = TV_k$. Подставив $V^* = TV^*$ и воспользовавшись сжатием,

$$\|V_{k+1} - V^*\|_\infty = \|TV_k - TV^*\|_\infty \le \gamma\,\|V_k - V^*\|_\infty,$$

то есть каждый шаг приближает оценку к $V^*$ хотя бы в $1/\gamma$ раз. Развернув эту рекуррентность до начального приближения, получаем геометрическую оценку

$$\|V_k - V^*\|_\infty \le \gamma^{\,k}\,\|V_0 - V^*\|_\infty.$$

Поскольку $\gamma < 1$, правая часть стремится к нулю, а вместе с ней и $\|V_k - V^*\|_\infty \to 0$. Таким образом, value iteration сходится к $V^*$.

### Asynchronious value iteration (2 pts)

Consider the following algorithm:

---

Initialize $V_0$

**for** $k = 0,1,2,...$ **do**

&nbsp;&nbsp;&nbsp;&nbsp; Select some state $s_k \in \mathcal{S}$    

&nbsp;&nbsp;&nbsp;&nbsp; $V(s_k) := (TV)(s_k)$

**end for**

---


Note that unlike common value iteration, here we update only a single state at a time.

**Homework.** Prove the following proposition:

If for all $s \in \mathcal{S}$, $s$ appears in the sequence $(s_0, s_1, ...)$ infinitely often, then $V$ converges to $V*$

**Доказательство**

Будем следить за величиной $\delta = \|V - V^*\|_\infty$ — наибольшим по состояниям отклонением текущей оценки от оптимума.

Назовём *проходом* такой отрезок итераций, на котором каждое состояние успевает обновиться хотя бы однажды. По условию любое состояние встречается в $(s_0, s_1, \ldots)$ бесконечно часто, поэтому проходы можно выделять один за другим сколь угодно долго.

Зафиксируем один проход и обозначим $\delta$ ошибку в его начале. Рассмотрим момент обновления отдельного состояния $s$: его новое значение равно $(TV)(s)$, и, так как $V^*$ неподвижно ($V^* = TV^*$), а $T$ сжимает,

$$|(TV)(s) - V^*(s)| = |(TV)(s) - (TV^*)(s)| \le \gamma\,\|V - V^*\|_\infty \le \gamma\,\delta.$$

Стало быть, сразу после обновления отклонение в $s$ не превосходит $\gamma\delta$. Состояния, которых обновление не коснулось, своих значений не меняют, и их ошибка не растёт — поэтому общее отклонение вдоль прохода не выходит за $\delta$, что и оправдывает оценку выше на каждом шаге. К концу прохода обновлёнными оказываются все состояния, и максимум их ошибок не превышает $\gamma\delta$.

Получается, один проход сжимает ошибку в $\gamma$ раз. После $m$ проходов

$$\|V - V^*\|_\infty \le \gamma^{\,m}\,\|V_0 - V^*\|_\infty \xrightarrow[m \to \infty]{} 0,$$

откуда $V \to V^*$. Итак, асинхронная value iteration сходится к оптимальной функции ценности $V^*$.

# HW Part 2: Policy iteration

## Policy iteration implementateion (3 pts)

Let's implement exact policy iteration (PI), which has the following pseudocode:

---
Initialize $\pi_0$   `// random or fixed action`

For $n=0, 1, 2, \dots$
- Compute the state-value function $V^{\pi_{n}}$
- Using $V^{\pi_{n}}$, compute the state-action-value function $Q^{\pi_{n}}$
- Compute new policy $\pi_{n+1}(s) = \operatorname*{argmax}_a Q^{\pi_{n}}(s,a)$
---

Unlike VI, policy iteration has to maintain a policy - chosen actions from all states - and estimate $V^{\pi_{n}}$ based on this policy. It only changes policy once values converged.


Below are a few helpers that you may or may not use in your implementation.

```python
transition_probs = {
    's0': {
        'a0': {'s0': 0.5, 's2': 0.5},
        'a1': {'s2': 1}
    },
    's1': {
        'a0': {'s0': 0.7, 's1': 0.1, 's2': 0.2},
        'a1': {'s1': 0.95, 's2': 0.05}
    },
    's2': {
        'a0': {'s0': 0.4, 's1': 0.6},
        'a1': {'s0': 0.3, 's1': 0.3, 's2': 0.4}
    }
}
rewards = {
    's1': {'a0': {'s0': +5}},
    's2': {'a1': {'s0': -1}}
}

from mdp import MDP
mdp = MDP(transition_probs, rewards, initial_state='s0')
```

Let's write a function called `compute_vpi` that computes the state-value function $V^{\pi}$ for an arbitrary policy $\pi$.

Unlike VI, this time you must find the exact solution, not just a single iteration.

Recall that $V^{\pi}$ satisfies the following linear equation:
$$V^{\pi}(s) = \sum_{s'} P(s,\pi(s),s')[ R(s,\pi(s),s') + \gamma V^{\pi}(s')]$$

You'll have to solve a linear system in your code. (Find an exact solution, e.g., with `np.linalg.solve`.)

```python
def compute_vpi(mdp, policy, gamma):
    """
    Computes V^pi(s) FOR ALL STATES under given policy.
    :param policy: a dict of currently chosen actions {s : a}
    :returns: a dict {state : V^pi(state) for all states}
    """
    states = list(mdp.get_all_states())
    n = len(states)
    idx = {s: i for i, s in enumerate(states)}

    # V = R + gamma * P * V  =>  (I - gamma*P) * V = R
    A = np.eye(n)
    b = np.zeros(n)

    for s in states:
        i = idx[s]
        if mdp.is_terminal(s):
            continue
        a = policy[s]
        for next_s, prob in mdp.get_next_states(s, a).items():
            r = mdp.get_reward(s, a, next_s)
            b[i] += prob * r
            A[i, idx[next_s]] -= gamma * prob

    v = np.linalg.solve(A, b)
    return {s: v[idx[s]] for s in states}
```

```python
test_policy = {s: np.random.choice(
    mdp.get_possible_actions(s)) for s in mdp.get_all_states()}
new_vpi = compute_vpi(mdp, test_policy, gamma)

print(new_vpi)

assert type(
    new_vpi) is dict, "compute_vpi must return a dict {state : V^pi(state) for all states}"
```

Once we've got new state values, it's time to update our policy.

```python
def compute_new_policy(mdp, vpi, gamma):
    """
    Computes new policy as argmax of state values
    :param vpi: a dict {state : V^pi(state) for all states}
    :returns: a dict {state : optimal action for all states}
    """
    return {s: get_optimal_action(mdp, vpi, s, gamma)
            for s in mdp.get_all_states()}
```

```python
new_policy = compute_new_policy(mdp, new_vpi, gamma)

print(new_policy)

assert type(
    new_policy) is dict, "compute_new_policy must return a dict {state : optimal action for all states}"
```

__Main loop__

```python
def policy_iteration(mdp, policy=None, gamma=0.9, num_iter=1000, min_difference=1e-5):
    """
    Run the policy iteration loop for num_iter iterations or till difference between V(s) is below min_difference.
    If policy is not given, initialize it at random.
    """
    if policy is None:
        policy = {s: (None if mdp.is_terminal(s) else np.random.choice(mdp.get_possible_actions(s)))
                  for s in mdp.get_all_states()}

    state_values = {s: 0 for s in mdp.get_all_states()}

    for i in range(num_iter):
        state_values = compute_vpi(mdp, policy, gamma)
        new_policy = compute_new_policy(mdp, state_values, gamma)
        new_state_values = compute_vpi(mdp, new_policy, gamma)

        diff = max(abs(new_state_values[s] - state_values[s])
                   for s in mdp.get_all_states())
        print("iter %4i   |   diff: %6.5f" % (i, diff))

        policy = new_policy
        state_values = new_state_values
        if diff < min_difference:
            break

    return state_values, policy
```

## Policy iteration convergence (3 pts)

**Note:** Assume that $\mathcal{S}, \mathcal{A}$ are finite.

We can define another Bellman operator:

$$(T_{\pi}V)(s) = \mathbb{E}_{r, s'|s, a = \pi(s)}\left[r + \gamma V(s')\right]$$

And rewrite policy iteration algorithm in operator form:


---

Initialize $\pi_0$

**for** $k = 0,1,2,...$ **do**

&nbsp;&nbsp;&nbsp;&nbsp; Solve $V_k = T_{\pi_k}V_k$   

&nbsp;&nbsp;&nbsp;&nbsp; Select $\pi_{k+1}$ s.t. $T_{\pi_{k+1}}V_k = TV_k$ 

**end for**

---

To prove convergence of the algorithm we need to prove two properties: contraction an monotonicity.

#### Monotonicity (0.5 pts)

For all $V, U$ if $V(s) \le U(s)$   $\forall s \in \mathcal{S}$ then $(T_\pi V)(s) \le (T_\pi U)(s)$   $\forall s \in  \mathcal{S}$

**Proof.**

Для произвольного состояния $s$

$$(T_\pi V)(s) = \sum_{s'} P(s' \mid s, \pi(s))\bigl[\,r(s, \pi(s), s') + \gamma V(s')\,\bigr].$$

Если $V(s') \le U(s')$ для всех $s'$, то, поскольку $\gamma \ge 0$,

$$r(s, \pi(s), s') + \gamma V(s') \le r(s, \pi(s), s') + \gamma U(s').$$

Все вероятности перехода неотрицательны, поэтому умножение на $P(s' \mid s, \pi(s))$ и суммирование по $s'$ сохраняют неравенство:

$$(T_\pi V)(s) \le (T_\pi U)(s).$$

Значит, оператор $T_\pi$ монотонен.

#### Contraction (1 pts)

$$
||T_\pi V - T_\pi U||_{\infty} \le \gamma ||V - U||_{\infty}
$$

For all $V, U$

**Proof.**

Оценим разность поточечно. При вычитании $T_\pi U$ из $T_\pi V$ награды сокращаются, и остаётся лишь дисконтированная разность ценностей:

$$|(T_\pi V)(s) - (T_\pi U)(s)| = \gamma\left|\sum_{s'} P(s' \mid s, \pi(s))\,\bigl(V(s') - U(s')\bigr)\right|.$$

Вносим модуль под сумму (неравенство треугольника) и заменяем каждое $|V(s') - U(s')|$ его супремумом $\|V - U\|_\infty$:

$$\le \gamma \sum_{s'} P(s' \mid s, \pi(s))\,|V(s') - U(s')| \le \gamma\,\|V - U\|_\infty \sum_{s'} P(s' \mid s, \pi(s)).$$

Оставшаяся сумма вероятностей равна единице, так что правая часть сводится к $\gamma\,\|V - U\|_\infty$. Эта оценка не зависит от $s$, и максимум по состояниям в левой части даёт

$$\|T_\pi V - T_\pi U\|_\infty \le \gamma\,\|V - U\|_\infty.$$

#### Convergence (1.5 pts)

Prove that there exists iteration $k_0$ such that $\pi_k = \pi^*$ for all $k \ge k_0$

**Proof.**

Разберём один шаг алгоритма. Текущая политика $\pi_k$ оценивается точно, поэтому $V^{\pi_k}$ — неподвижная точка $T_{\pi_k}$. Очередная политика $\pi_{k+1}$ выбирается жадно, то есть так, что $T_{\pi_{k+1}} V^{\pi_k} = T V^{\pi_k}$.

Сначала покажем, что ценность не убывает. Оператор $T$ максимизирует по действиям, а $\pi_k$ — лишь один из возможных выборов, поэтому

$$T_{\pi_{k+1}} V^{\pi_k} = T V^{\pi_k} \ge T_{\pi_k} V^{\pi_k} = V^{\pi_k}.$$

Применяя к неравенству $T_{\pi_{k+1}} V^{\pi_k} \ge V^{\pi_k}$ оператор $T_{\pi_{k+1}}$ многократно и пользуясь его монотонностью, получаем $T_{\pi_{k+1}}^m V^{\pi_k} \ge V^{\pi_k}$ при всех $m$. Левая часть при $m \to \infty$ сходится к неподвижной точке $T_{\pi_{k+1}}$, то есть к $V^{\pi_{k+1}}$, и в пределе

$$V^{\pi_{k+1}} \ge V^{\pi_k}.$$

Значит, ценности образуют неубывающую последовательность.

Теперь — почему процесс конечен. Каждый раз, когда политика действительно меняется, её ценность строго возрастает хотя бы в одном состоянии; единственное исключение — когда $\pi_k$ уже жадна относительно собственной ценности, и менять попросту нечего. Детерминированных политик конечное число, и каждой отвечает своя ценность, поэтому строгий рост не может повторяться бесконечно. Следовательно, существует номер $k_0$, с которого политика перестаёт меняться.

Остаётся проверить, что стабильная политика оптимальна. Её стабильность означает жадность по своей же ценности, $T_{\pi_{k_0}} V^{\pi_{k_0}} = T V^{\pi_{k_0}}$, а вместе с определением $V^{\pi_{k_0}} = T_{\pi_{k_0}} V^{\pi_{k_0}}$ это даёт

$$V^{\pi_{k_0}} = T V^{\pi_{k_0}}.$$

Перед нами уравнение оптимальности Беллмана, единственное решение которого — $V^*$. Поэтому $V^{\pi_{k_0}} = V^*$, политика $\pi_{k_0}$ оптимальна, а при фиксированном правиле разрешения ничьих та же политика выбирается и на всех дальнейших итерациях. Итак, $\pi_k = \pi^*$ для всех $k \ge k_0$.
